2. (25 points) The file restaurant_reviews.txt is given to you. 
 
The file contains names of reviewers and their reviews of different restaurants in the comma separated format. The first string is the name of the reviewer followed by a name of a restaurant and its rating. You are required to write a Python program that computes a similarity score between any two reviewers using Euclidean distances. An example data from the restaurant_reviews.txt file is given below: 


Solution / Read me:
1. Program begins by prompting user for file to read, exception for file not found handled.
2. Upon right fill name, file is read and split based on comma delimiter, used a temporary dictionary to store each comma delimited values by running through a loop and finally assigining reviewers name as key and the values from the temp dictionary as values. 
4. User is prompted between 3 choices (1,2,3) 
5. On selecting 1, User can enter two reviewers name to be compared, exception for invalid names handled. 
6. Since reviewer name is key in dicitionary, it has to be exactly as it is stored in dictionary. 
7. Temporary dictionary of the two name from user is created where restaraunt name and rating is stored as key value pair
8. A list of common restaurant name (key) is obtained and the euclidean score is calculated for all the matching keys.
9. Score is shown with 4 decimal value
10. On selecting 2, User enter 1 reviewers name to be compared and scored against all other reviwers(key) in the database. 
11. Similar logic is followed for finding the common rated restaraunt. 
12. loop is maintained for computing the comparative scores between user choice of reviwer and all others in database.
13. A key of "\n" is stored and results in a unique key-value pair in the main dictionary and yields a result of 0.0 in the last row. 

```python
import random
import math

def main():
    while True:
        try:
            filename = input("Give the name of the restaurant reviews file: ")
            file = open(filename)
            break
        except FileNotFoundError:
            print("File not found, Please check the filename again")
            continue        
    dictReview = dict()
    for line in file:
        listDetails = line.split(',')
    #temporary dictionary to assign key (reviewer name) with all details as value
        dictTemp = dict()
        size  = len(listDetails)
    #print(size)
        i = 1
    #iterate through to add all values from each line 
        while i+1 < size:
            dictTemp[listDetails[i]]=listDetails[i+1]
            i = i+2
        #print(dictTemp)
    #assign reviewers name as key    
        dictReview[listDetails[0]] = dictTemp
#print(dictReview)
    choice = 0
    while(choice == 0):
            userInput = input("What do you want to do? Input 1 for similarity between two reviewers,or Input 2 for similarity between one reviewer and all others in the database or 3 to quit:")
            if(userInput == '1'):
                while True:
                    nameOne = input("Provide Reviewer1 name: ")
                    nameTwo = input("Provide Reviewer2 name: ")
                    flagName1 = (nameOne not in dictReview.keys())
                    flagName2 = (nameTwo not in dictReview.keys())
                    if(flagName1 or flagName2):
                        print("Please enter a reviewer name in the database")
                        continue
                    #print("Success names")
                    #print(flagName1)
                    #print(flagName2)
                    else:
                        score = 0
                        #get values for all restaraunts for reviewer1, key - restaraunt, value - rating
                        valueOne = dictReview[nameOne]
                        #get values for all restaraunts for reviewer2, key - restaraunt, value - rating 
                        valueTwo = dictReview[nameTwo]
                        #store keys
                        restReviewed1 = set(valueOne.keys())
                        restReviewed2 = set(valueTwo.keys())
                        #get the common restaraunts that both have reviewed
                        restCommon = list(restReviewed1 & restReviewed2)
                        #print(valueOne)
                        #print(valueTwo)
                        #print(restReviewed1)
                        #print(restReviewed2)
                        #print(restCommon)
                        for key in restCommon:
                            score += round(pow((float(valueOne[key])-float(valueTwo[key])),2),4)
                            #print(score)
                        print("The similarity score between ",nameOne," and ",nameTwo,"  is: ",round(math.sqrt(score),4))
                    break    
            elif(userInput == '2'):
                 while True:
                    reviewer = input("Provide Reviewer name: ")
                    flagName = (reviewer not in dictReview.keys())
                    if(flagName):
                        print("Please enter a reviewer name in the database")
                        continue
                    else:
                        print("The Similarity Scores are: ")
                        for name in list(dictReview.keys()):
                            if name != reviewer:
                                comparisonScore = 0
                                otherReviewers = name
                                reviewerValue = dictReview[reviewer]
                                othersValue = dictReview[otherReviewers]
                                reviewerKey = set(reviewerValue.keys())
                                othersKey = set(othersValue.keys())
                                commonKey = list(reviewerKey & othersKey)
                                #print(reviewerValue)
                                #print(othersValue)
                                #print(reviewerKey)
                                #print(othersKey)
                                #print(commonKey)
                                for key in commonKey:
                                    comparisonScore += pow((float(reviewerValue[key])-float(othersValue[key])),2)
                                    #print(comparisonScore)
                                print(reviewer,"               ",otherReviewers,"             ",round(math.sqrt(comparisonScore),2))
                    break            
            elif(userInput == '3'):
                choice = 1
                break
            else:
                print("Invalid option, please select from one of the given options")
                continue
    print("Good bye!")
                
main()
```
